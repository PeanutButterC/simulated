#### Each trajectory file contains two hdf5 dataframes, the *indices* and *trajectories*.

The **indices** refer to the unique identifier of each session.
The **trajectories** contain a 10801x11 matrix of trajectory and state information. The rows are timesteps. The columns are the following:

- target pos X
- target pos Y
- target pos Z
- target rot X
- target rot Y
- target rot Z
- target rot W
- hand pos X
- hand pos Y
- hand pos Z
- agent state

```python
with h5py.File(datapath, 'r') as f:
    indices = f['indices'][:] # session IDs
    trajectories = f['df'][:,:,:-1] # trajectories, ignoring the last column (agent state)
```
